```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("D:Maged/imdb/imdb_md.csv")
data.head()
```

```python
data.info()
```

```python
data.drop(columns={'id','imdb_id','homepage','keywords','tagline','original_title'},inplace=True)
# drop uncesssary columns in the dataset
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
ind=data[data['runtime']==0].index
data.drop(ind,axis=0,inplace=True)
data[data['runtime']==0]
#droping movies without a specific runtime "runtime=0".
```

```python
data['release_date']=pd.to_datetime(data['release_date'])
# Turning "release_date" from "string" to "datetime".
# To be easier to search by date.
```

```python
data.sample(10)
```

```python
data.genres.mode()
# Which genres are most popular in this dataset.
```

```python
data.production_companies.mode()
# Which production company has more movies than the other production companies in this dataset.
```

```python
data.director.mode()
# Which director has more movies than the other directors in this dataset.
```

```python
data.cast.mode()
# Which actor/actress has more movies than the other actors/actresses in this dataset.
```

```python
data.release_year.mode()
# Which year has more movies than the other years in this dataset.
```

```python
data.release_date.dt.month.mode()
# What is the most popular release month
# The most popular month is >>>> September
```

```python
high_revenue=data.sort_values(by=['revenue'],ascending=False).head(10)
high_revenue.genres.mode()
# Finding movies genres that have the highest revenues. The most popular movies genres are "Action|Adventure|Science Fiction".
```
